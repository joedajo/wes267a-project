#warning_logic.ipynb

notebook contains threshold checking functions for the Receiver PYNQ board.
- temperature thresholds to detect *HOT* or *COLD* conditions.
- mic signal levels to detect *LOUD* environmental noise.
The output of these functions is a concise status string (e.g., `NORMAL`, `HOT`, `COLD`, `LOUD`, or combinations such as `HOT/LOUD`).  
This status is used by the receiver system to:
- display warnings in the Jupyter notebook console.
- send warning messages to the oled display for real-time system monitoring.

```python
def evaluate_status(temp_c, mic_level,
                    hot_temp_c=30.0,
                    cold_temp_c=15.0,
                    loud_noise=500.0):
  
    warnings = []

    if temp_c >= hot_temp_c:
        warnings.append("Too Hot!")
    elif temp_c <= cold_temp_c:
        warnings.append("Too Cold!")

    if mic_level >= loud_noise:
        warnings.append("Too Loud!")

    if not warnings:
        return "NORMAL"

    return "|".join(warnings)
```

```python
def combine_status(derived_status, temp_status, mic_status):
 
    status_parts = []

    if derived_status and derived_status != "NORMAL":
        status_parts.append(derived_status)

    if temp_status and temp_status != "NORMAL":
        status_parts.append(temp_status)

    if mic_status and mic_status != "NORMAL":
        status_parts.append(mic_status)

    if not status_parts:
        return "NORMAL"

    # remove duplicates while preserving order
    final_parts = []
    for item in status_parts:
        if item not in final_parts:
            final_parts.append(item)

    return "|".join(final_parts)
```

```python
# Optional quick test cell
print(evaluate_status(22.0, 200.0))
print(evaluate_status(35.0, 700.0))
print(combine_status("HOT/LOUD", "HOT", "LOUD"))
```
